1.3.1 Задание: Создать 8x8 матрицу и заполнить её в шахматном порядке нулями и
единицами.

```python
import numpy as np

matrix = np.zeros((8, 8), dtype=int)
matrix[1::2, ::2] = 1
matrix[::2, 1::2] = 1

print(matrix)
```

1.3.4 Задание: Создать матрицу с 0 внутри, и 1 на границах.

```python
import numpy as np

n = 8
x = np.zeros((n, n), dtype=int)
for i in range(n):
  if i==0 or i==(n-1):
    x[i] = np.array([1]*n)
  else:
    for j in range(n):
      if j ==0 or j==(n-1):
        x[i][j] = 1

print(x)
```

1.3.6 Задание: Создайте матрицу, выведите ее форму, размер и размерность.

```python
import random
from random import randint
import numpy as np

x = np.random.random((randint(0, 10), randint(0, 10)))
print(x)
dim = np.shape(x)
if dim[0] == dim[1]:
  print("Квадратная")
else:
  print("Прямоугольная")
print("Размер: ", np.size(x))
print("Размерность: ", dim)
```

3.3.2 Задание: Загрузить фрейм данных по ссылке:
https://raw.githubusercontent.com/akmand/datasets/master/iris.csv.
Необходимо выполнить нормализацию первого числового признака
(sepal_length_cm) с использованием минимаксного преобразования, а
второго (sepal_width_cm) с задействованием z-масштабирования

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler

dataframe = pd.read_csv('https://raw.githubusercontent.com/akmand/datasets/master/iris.csv')

scaler_minmax = MinMaxScaler()
dataframe['sepal_length_cm'] = scaler_minmax.fit_transform(dataframe[['sepal_length_cm']])

scaler_zscore = StandardScaler()
dataframe['sepal_width_cm'] = scaler_zscore.fit_transform(dataframe[['sepal_width_cm']])

print(dataframe)
```
